LIMPIEZA DE DFs

```python
#Importacion de pandas
import pandas as pd
# Obtenci√≥n de la tabla de Excel sin encabezado
data1 = pd.read_excel(file1, header=None)

#diferenciar HDI rank 2021 y 2020
data1.loc[4, 14] = "HDI rank 2020"

data1.head(9)
```

```python
#seleccionar el encabezado
encabezado=data1.iloc[4]
data1.columns=encabezado
# Eliminar las filas iniciales
data1=data1.drop(index=[0,1,2,3,4,5,6])

# Restablecer el √≠ndice
data1.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame resultante
data1
```

```python
# Eliminar las primeras 4 filas
data_1= data1.drop(index=[0, 1, 2, 3])

# Restablecer el √≠ndice
data1.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame resultante
data_1
```

```python
#Eliminacion de todas las filas que contengan valores NaN
data1_clean= data1[~(data1.iloc[:, 1:].isna().all(axis=1) & data1.iloc[:, 0].notna())]

data1_clean.head()
```

```python
# Eliminaci√≥n de columnas que contienen NaN en la primera fila
data1_clean = data1_clean.loc[:, data1_clean.iloc[0].notna()]


# Reiniciar el √≠ndice del DataFrame
data1_clean.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame resultante
data1_clean.head(5)
```

```python
data1_clean.info()
```

```python
# Convertir las columnas de la 2 a la 7 a n√∫meros
for col in data1_clean.columns[2:8]:
    data1_clean[col] = pd.to_numeric(data1_clean[col], errors='coerce')  # Convertir a float64

# Verificar el resultado
data1_clean
```

```python
data1_clean.info()
```

```python
#Limpieza file4
data4 = pd.read_excel(file4, header=None)
encabezado=data4.iloc[0]
data4.columns=encabezado
data4 = data4.drop(index=0)
data4.reset_index(drop=True, inplace=True)


data4.head(7)
```

```python
data4_clean= data4[~(data4.iloc[:, 2:].isna().all(axis=1) & data4.iloc[:, 1].notna())]  #eliminar filas con NaN
data4_clean = data4_clean.loc[:, data4_clean.iloc[0].notna()]   #eliminar columna con NaN
data4_clean.reset_index(drop=True, inplace=True)
data4_clean.head()
```

```python
for col in data4_clean.columns[3:9]:
    data4_clean[col] = pd.to_numeric(data4_clean[col], errors='coerce')
data4_clean.head()
data4_clean.info()
```

```python
#limpieza file2
data2 = pd.read_excel(file2, header=None)
data2.loc[0, 11] = "HDI rank 2020"
encabezado=data2.iloc[0]
data2.columns=encabezado

data2= data2.drop(index=[0, 1, 2])
data2.reset_index(drop=True, inplace=True)

data2_clean= data2[~(data2.iloc[:, 1:].isna().all(axis=1) & data2.iloc[:, 0].notna())]  #eliminar filas con NaN
data2_clean = data2_clean.loc[:, data2_clean.iloc[0].notna()]   #eliminar columna con NaN
data2_clean = data2_clean.drop(columns=data2.columns[6])   #tiene valores adicionales al NaN
data2_clean.reset_index(drop=True, inplace=True)

for col in data2_clean.columns[2:8]:
    data2_clean[col] = pd.to_numeric(data2_clean[col], errors='coerce')

data2_clean.head()

data2_clean.info()
```

```python
# limpieza file 3
data3 = pd.read_excel(file3, header=None)
data3.loc[0, 14] = "HDI rank 2020"
encabezado=data3.iloc[0]
data3.columns=encabezado

data3= data3.drop(index=[0, 1, 2])
data3.reset_index(drop=True, inplace=True)

data3_clean= data3[~(data3.iloc[:, 1:].isna().all(axis=1) & data3.iloc[:, 0].notna())]  #eliminar filas con NaN
data3_clean = data3_clean.loc[:, data3_clean.iloc[0].notna()]   #eliminar columna con NaN
data3_clean.reset_index(drop=True, inplace=True)

for col in data3_clean.columns[2:9]:
    data3_clean[col] = pd.to_numeric(data3_clean[col], errors='coerce')



#data3_clean = data3_clean.drop(columns=data3.columns[8])

data3_clean.head()
```

INTEGRACI√ìN DE DFs

```python
# creaci√≥n de lista de los 4 data frames
import pandas as pd

DFs=[data1_clean,data2_clean,data3_clean,data4_clean]
```

```python
# guardando los nombres de las columnas
allColumnNames=[]
for df in DFs:
    allColumnNames.append(set(df.columns))

# verificando si tienen la misma cantidad
[len(cols) for cols in allColumnNames]
```

```python
# detalle de columnas en com√∫n
commonColumns=set.intersection(*allColumnNames)
len(commonColumns),commonColumns
```

```python
# columna diferente
set.union(*allColumnNames)-commonColumns
```

```python
#eliminar esa columna
allDFs_sameNames=[df.loc[:,list(commonColumns)] for df in DFs]
len(allDFs_sameNames)
```

```python
#concatenar
#allDFs_sameNames=[df.loc[:,list(commonColumns)] for df in DFs]
allDFsConcat = pd.concat(allDFs_sameNames, ignore_index=True)
allDFsConcat
```

```python
allDFsConcat.info()
```

```python
#limpieza de nombres de paises
# ubicar paises con nombres varios
allDFsConcat['Country'] = allDFsConcat['Country'].str.replace(r'\([^)]*\)', '', regex=True)
allDFsConcat['Country'] = allDFsConcat['Country'].str.split(',').str[0]
allDFsConcat['Country'] = allDFsConcat['Country'].str.upper()
allDFsConcat.head(5)
```

MERGING

```python
link1="https://github.com/Lorbe277/Control3_ICDE/raw/main/countryCODES.xlsx"

import pandas as pd
# Tabla CountryCODES
country=pd.read_excel(link1, header=None)
country
```

```python
encabezado1=country.iloc[0]
country.columns=encabezado1
country
```

```python
# merge de ambas DFs
merge=allDFsConcat.merge(country,left_on='Country',right_on='Country')
merge
```
